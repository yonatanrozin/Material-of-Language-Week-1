```python
import textwrap
myText = open('newlinewiki.txt').read()
wrapText = textwrap.fill(myText,88)
print(wrapText)
```

```python
rows = wrapText.split("\n") #list of rows
numRows = len(rows)
newRows = []
for letter in range(70):
    rotatedRow = '' #create a new row
    for n in range(numRows-1): #take the nth character of each row
        rotatedRow+=rows[n][letter]+" "
    newRows.append(rotatedRow)
for i in range(len(newRows)):
    print(newRows[i])
    
        
```
